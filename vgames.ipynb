Crearemos una función para convertir los nombres de columnas del DataFrame de camelCase a snake_case la cual aplicaremos a todos los data frames.

```python
_underscorer1 = re.compile(r'(.)([A-Z][a-z]+)')
_underscorer2 = re.compile('([a-z0-9])([A-Z])')

def to_snake_case(columns):
    
    return list(map(lambda string: _underscorer2.sub(r'\1\2', _underscorer1.sub(r'\1\2', string)).lower(), columns))
```

### metagaming

Importaremos esta tabla de la web, para así poder imputar muchos de los valores ausentes, por medio de una función.

```python
url = 'https://raw.githubusercontent.com/graflamerich/videogames/main/VG-YEARS-22.csv'
```

```python
new_data =pd.read_csv(url, encoding='ISO-8859-1')
```

```python
new_data.info()
```

```python
new_data.head()
```

```python
pivot = new_data.pivot_table(index = ['name', 'platform'], values= 'year_of_release')
pivot.head()
```

```python
def years(row):
    name = row['name']
    platform = row['platform']
    year = row['year_of_release']
    try:
        if pd.isna(year):
            return pivot['year_of_release'][name][platform]
        return year
    except:
        return np.nan
```

```python
years(data.iloc[183])
```

```python
data['year_of_release'] = data.apply(years, axis=1)
```

La información en los datos que tenemos, es la sugerida por la **ESRB (Entertainment Software Rating Board)** empleada en Estados Unidos y consiste en clasificar los videojuegos según su contenido de violencia, sexo, lenguaje, temas, entre otros.

Los símbolos empleados son:

- **EC:** Early Childhood o "Infancia temprana", sugerido a partir de 3 años.
- **E:** Everyone o "Para todos": temas sugeridos personas de 6 años en adelante. 
- **E10+** Everyone 10 and older o "Para todos mayores de 10 años".
- **T:** (Teen o "Adolescentes"): para edades de 13 años o más.
- **K-A:** (Kids to Adult): para el final de la adolescencia que son niños en vias de adultos.
- **M:** (Mature o "Maduro"): el contenido es sugerido para 17 o más años.
- **AO:** (Adults Only o "Sólo Adultos"): contenido con escenas amplias de violencia, sexo o racismo, se sugiere únicamente para adultos.
- **RP:** (Rating Pending o "Clasificación Pendiente"): la llevan productos que han sido sometidos a evaluación ante la ESRB y están en espera de una categorización final.

En cuanto al rating y sus valores ausentes, al ser desconocido el valor, lo unificaremos al rating de RP que significa que el está pendiente de clasificación.

### XBox One user score VS. PC user score 

Nuestras hipótesis a plantear serán:

$H_0: \mu_{\text{vg}} = \mu_{\text{vc}}$

$H_a: \mu_{\text{vg}} \neq \mu_{\text{vc}}$

Establezcamos un nivel de significancia del 5%.

Levene False varianza menor 0.05 alpha y mayor True

Hipótesis nula  𝐻0 : Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

Hipótesis nula  𝐻𝑎 : Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.

```python
xbox_final['user_score'].dropna().mean()
```

```python
pc_final['user_score'].dropna().mean()
```

```python
stat, p_value = levene(xbox_final['user_score'].dropna(), pc_final['user_score'].dropna())
p_value
```

```python
alpha = 0.05
results = st.ttest_ind( xbox_final['user_score'].dropna(), pc_final['user_score'].dropna(), equal_var = True)
print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula") 
```

Evidencia suficiente, dado el nivel de significación que seleccionamos, para no rechazar la hipótesis nula. Por lo tanto, podemos concluir que las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son iguales.
